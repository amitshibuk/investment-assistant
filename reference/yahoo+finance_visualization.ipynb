```python
import yfinance as yf
import plotly.express as px

# List of tickers
tickers = ["AAPL", "MSFT", "GOOGL", "AMZN", "TSLA"]

for ticker in tickers:
    # Download data
    data = yf.download(ticker, start="2023-01-01", end="2025-01-01", auto_adjust=False)

    # Make sure y-axis is a 1D series
    close_prices = data['Close'].squeeze()

    # Create interactive plot
    fig = px.line(
        x=data.index,
        y=close_prices,
        title=f'{ticker} Closing Stock Price',
        labels={'x': 'Date', 'y': 'Closing Price (USD)'}
    )

    fig.show()
```
