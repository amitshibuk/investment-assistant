```python
import yfinance as yf
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt

# -----------------------------
# 1. Fetch historical data
# -----------------------------
symbol = "AAPL"
data = yf.download(symbol, start="2018-01-01", end="2025-01-01")

# Use only the 'Close' column
close_prices = data[['Close']]

# Scale values between 0 and 1
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(close_prices)

# -----------------------------
# 2. Train/Test Split
# -----------------------------
training_size = int(len(scaled_data) * 0.8)
train_data = scaled_data[:training_size]
test_data = scaled_data[training_size:]

# Function to create dataset
def create_dataset(dataset, time_step=60):
    X, y = [], []
    for i in range(len(dataset) - time_step - 1):
        X.append(dataset[i:(i + time_step), 0])
        y.append(dataset[i + time_step, 0])
    return np.array(X), np.array(y)

time_step = 60
X_train, y_train = create_dataset(train_data, time_step)
X_test, y_test = create_dataset(test_data, time_step)

# Reshape for LSTM [samples, time_steps, features]
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

# -----------------------------
# 3. Build LSTM Model
# -----------------------------
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))

model.compile(optimizer="adam", loss="mean_squared_error")
model.fit(X_train, y_train, batch_size=32, epochs=20, verbose=1)

# -----------------------------
# 4. Predictions
# -----------------------------
y_pred = model.predict(X_test)

# Inverse scale
y_predicted = scaler.inverse_transform(y_pred)
y_actual = scaler.inverse_transform(y_test.reshape(-1, 1))

# Metrics
mse = mean_squared_error(y_actual, y_predicted)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_actual, y_predicted)

print("MSE:", mse)
print("RMSE:", rmse)
print("MAE:", mae)

# -----------------------------
# 5. Future Predictions
# -----------------------------
last_60 = scaled_data[-time_step:]

def predict_future_days(model, last_sequence, days_ahead=7):
    future_predictions = []
    current_sequence = last_sequence.copy()

    for _ in range(days_ahead):
        X_future = current_sequence.reshape(1, time_step, 1)
        next_price = model.predict(X_future, verbose=0)
        future_predictions.append(next_price[0, 0])

        # Update sequence (append new value, drop first)
        current_sequence = np.append(current_sequence[1:], next_price).reshape(time_step, 1)

    return scaler.inverse_transform(np.array(future_predictions).reshape(-1, 1))

# Tomorrow
future_1d = predict_future_days(model, last_60, days_ahead=1)

# Next week (7 days)
future_7d = predict_future_days(model, last_60, days_ahead=7)

# Next month (30 days)
future_30d = predict_future_days(model, last_60, days_ahead=30)

print("\nüìÖ Predicted Tomorrow‚Äôs Price:", future_1d[-1][0])
print("üìÖ Predicted Next Week Prices:\n", future_7d.flatten())
print("üìÖ Predicted Next Month Prices:\n", future_30d.flatten())

# -----------------------------
# 6. Visualization (last actual vs next month forecast)
# -----------------------------

plt.figure(figsize=(12,6))
plt.plot(y_actual[-100:], label="Actual Price (last 100 days)")
plt.plot(y_predicted[-100:], label="Predicted Price (last 100 days)")
plt.plot(range(len(y_actual), len(y_actual)+30), future_30d, label="Next Month Prediction", linestyle="dashed")
plt.title("AAPL Price Prediction")
plt.xlabel("Days")
plt.ylabel("Price")
plt.legend()
plt.show()
```
